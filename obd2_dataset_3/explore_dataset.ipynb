```python
import pandas as pd
import numpy as np

df = pd.read_csv('Driving Data(KIA SOUL)_(150728-160714)_(10 Drivers_A-J).csv')
letter_to_int = {chr(ord('A') + i): i + 1 for i in range(10)}
df['Class'] = df['Class'].map(letter_to_int).astype(int)
trip_starts = df['Time(s)'] == 1
df['trip_number'] = trip_starts.cumsum()
df = df.drop(columns=['Time(s)'])
df
```

```python
df.to_csv('dataset_2.csv', index=False)
```
