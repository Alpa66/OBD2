```python
import pandas as pd

# Specify the delimiter as semicolon
df = pd.read_csv('joined.csv', delimiter=';', parse_dates=['datetime'])
df['datetime'] = df['datetime'].dt.floor('S')
display(df)
```

```python
df2 = pd.read_excel('_Бортовой журнал (электронный).xlsx')

df2 = df2[['Unnamed: 1', 'Unnamed: 2', ' - забор пробы масла']]
df2.columns = ['date', 'start_time', 'end_time']
df2 = df2.iloc[2:]
display(df2)
```

```python
df2['trip_start'] = pd.to_datetime(df2['date'].astype(str) + ' ' + df2['start_time'].astype(str), errors='coerce')
df2['trip_end'] = pd.to_datetime(df2['date'].astype(str) + ' ' + df2['end_time'].astype(str), errors='coerce')

df2 = df2[['trip_start', 'trip_end']]
```

```python
df2.info()
```

```python
df2['trip_start_in_df'] = df2['trip_start'].isin(df['datetime'])
df2['trip_end_in_df'] = df2['trip_end'].isin(df['datetime'])
display(df2[['trip_start', 'trip_start_in_df', 'trip_end', 'trip_end_in_df']])
```

```python
def is_in_any_interval(dt, intervals):
    # intervals is a DataFrame with 'trip_start' and 'trip_end'
    return ((intervals['trip_start'] <= dt) & (dt <= intervals['trip_end'])).any()

from tqdm import tqdm

# Register tqdm with pandas apply
tqdm.pandas()

df['in_trip'] = df['datetime'].progress_apply(lambda x: is_in_any_interval(x, df2))
display(df[['datetime', 'in_trip']])
```

```python
# Find the start and end indexes of continuous segments where 'in_trip' is False
false_mask = ~df['in_trip'].fillna(False).values

segments = []
start = None

for i, val in enumerate(false_mask):
    if val and start is None:
        start = i
    elif not val and start is not None:
        segments.append((start, i - 1))
        start = None

# Handle case where the last segment goes to the end
if start is not None:
    segments.append((start, len(false_mask) - 1))

# Display the segments
for seg_start, seg_end in segments:
    print(f"False segment: start index = {seg_start}, end index = {seg_end}")
```

```python
# Save the current dataframe to a CSV file
df.to_csv('joined_with_in_trip.csv', index=False, sep=';')
```
