```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/in/oil_protokol/oil_lab_protokol.csv', sep=';', decimal=',')
df.head()
```

```python
# Display basic info about the dataframe
print(df.info())

# Show the first few rows
display(df.head())

# Show summary statistics for numeric columns
display(df.describe())

# Show unique values for each column (helpful for categorical data)
for col in df.columns:
    print(f"Column: {col}")
    print(df[col].unique()[:10])  # Show up to 10 unique values
    print('-' * 40)
```

```python
# Drop columns where all values are zero
df = df.dropna(axis=0, how='all')
df = df.loc[:, ~(df == 0).all()]
display(df.head())
```
